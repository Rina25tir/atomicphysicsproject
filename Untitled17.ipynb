## Project: Parallel implementation of high-fidelity multiqubit gates with neutral atoms
The probability of an atom being in state |1> is given by:
$$ P_1(t) = sin^2(\frac{\Omega_{01}t}{2}), $$

where $\Omega_{01}$ is the effective Rabi frequency of transitions between state $|0>$ and $|1>$. In this problem it
has a value of approximately $2\pi \ x \ 250 \ kHz$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
omega_01 = 2 * np.pi * 250e3  # Effective Rabi frequency in Hz (250 kHz)
time_microseconds = np.linspace(0, 20, 1000)  # Time in microseconds
time_seconds = time_microseconds * 1e-6  # Convert time to seconds

# Rabi oscillation formula
P_1 = np.sin(omega_01 * time_seconds / 2)**2

# Plot
plt.figure(figsize=(8, 5))
plt.plot(time_microseconds, P_1, color='violet',label="Rabi Oscillations")
plt.title("Rabi Oscillations: Probability of |1 vs. Raman Pulse Time", fontsize=14)
plt.xlabel("Raman Pulse Time (s)", fontsize=12)
plt.ylabel("P", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

Similarly, I show here the Rydberg Rabi oscillations from $|1>$ to Rydberg state $|r>$. Only one atom in each pair is prepared in state |1> to avoid interactions. The effective Rabi frequency $\Omega$ is approximately $2\pi \ x \ 3.5 \ MHz$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the Rydberg Rabi oscillations
omega_rydberg = 2 * np.pi * 3.5e6  # Effective Rabi frequency (3.5 MHz)
time_microseconds = np.linspace(0, 2, 1000)  # Time in microseconds
time_seconds = time_microseconds * 1e-6  # Convert time to seconds

# Calculate Rabi oscillations (probability of |r state)
P_r = np.sin(omega_rydberg * time_seconds / 2)**2

# Plot Rydberg Rabi oscillations
plt.figure(figsize=(8, 5))
plt.plot(time_microseconds, P_r, color='green', label=r"$P_r = \sin^2(\frac{\Omega t}{2})$")
plt.title("Rydberg Rabi Oscillations: Probability of |r vs. Pulse Time", fontsize=14)
plt.xlabel("Rydberg Pulse Time (s)", fontsize=12)
plt.ylabel("P", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
